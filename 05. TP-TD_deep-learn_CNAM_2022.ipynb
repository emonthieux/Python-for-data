# Deep learning, réseaux de neurones

Le Deep Learning ou apprentissage profond est une technique de machine learning reposant sur le modèle des réseaux neurones. Plusieurs couches de neurones sont empilées pour apporter une plus grande complexité à l’établissement des règles.

![gif_deep-learning_dog](https://miro.medium.com/max/1400/1*-eLjPY7UGSoQhSyW5qC6gw.gif)

Les réseaux de neurones (en anglais neural networks) sont une classe de modèles qui sont construits à l'aide de couches de neurones. Les réseaux de neurones convolutionnels (en anglais convolutional neural networks) ainsi que les réseaux de neurones récurrents (en anglais recurrent neural networks) font parti des principaux types de réseaux de neurones.

Le Deep Learning (apprentissage profond) s'appuie sur un réseau de neurones artificiels s'inspirant du cerveau humain. Ce réseau est composé d'une multitude de « couches » de neurones, chacune recevant et interprétant les informations de la couche précédente. Ces techniques ont permis des progrès importants et rapides dans les domaines de l'analyse du signal sonore ou visuel et notamment de la reconnaissance faciale, de la reconnaissance vocale, de la vision par ordinateur, du traitement automatique du langage.

## Concepts fondamentaux - quels sont les composants ?

### Les neurones :
Les neurones correspondant à l'ensemble de fonctions mathématiques qui prennent une donnée d’entrée et produisent une donnée de sortie. Plusieurs neurones sont regroupés en couches aussi appelées *layers*. 

Les neurones d’entrée reçoivent des données d’entrée, les traitent et les transmettent aux neurones dans la couche suivante. Les neurones cachés prennent les données de sortie des précédents neurones en entrée, calculent de nouvelles données de sortie et les transmettent à des couches successives.


### Les couches
Les couches, aussi appelées *layers* contiennent plusieurs neurones (indiqués par l'argument `units =` dans la création du réseau) et aident à faire circuler l’information. Il existe au moins deux couches dans un réseau de neurones: la couche d’entrée (input layer) et la couche de sortie (output layer). En plus de ces deux couches requise, les réseaux de neurones contiennent (la plus part du temps) de nombreuses couches dites cachées. Les couches cachées correspondent donc aux couches de neurones comprises entre la couche d'entrée et la couche de sortie.

Ces couches cachées correspondent donc à la **_"black box"_** dont on parle souvent en cas de deep learning.

### Les poids et biais
Les poids et biais sont des variables du modèle qui sont mises à jour pour améliorer la précision du réseau. Un poids est appliqué à l’entrée de chacun des neurones pour calculer une donnée de sortie.
Les réseaux de neurones mettent à jour ces poids de manière continue. Il existe donc une boucle de rétro-action mise en œuvre dans la plupart des réseaux de neurones.

Les biais sont également des valeurs numériques qui sont ajoutées une fois que les poids sont appliqués aux valeurs d’entrée. Les poids et les biais sont donc en quelque sorte des valeurs d’auto-apprentissage de nos réseaux de neurones.


### La fonction d'activation
Les fonctions d’activation lissent ou normalisent la donnée de sortie avant qu’elle ne soit transmise aux neurones suivants. Ces fonctions aident les réseaux de neurones à apprendre et à s’améliorer.

Dit simplement, la fonction d’activation permet de changer notre manière de voir une donnée.
Par exemple si on a comme donnée : chaque semaine 50% des clients d’un magasin achètent des barres de chocolat; la fonction d’activation permettrait de changer la donnée en 50% des clients aiment le chocolat ou encore, 50% des clients prévoient d’acheter du chocolat chaque semaine.

Il s'agit d'une formule mathématique activée dans certaines circonstances. Lorsque les neurones calculent la somme pondérée des valeurs d’entrée + le biais, elles sont transmises à la fonction d’activation, qui vérifie si la valeur calculée est supérieure au seuil requis. Si c'est bien le cas, la fonction d’activation est activée et une valeur de sortie est calculée. Cette valeur de sortie est ensuite transmise aux couches suivantes ou précédentes (en fonction de la complexité du réseau), ce qui peut aider les réseaux de neurones à modifier le poids de leurs neurones.

Les fonctions d’activation introduisent la non-linéarité dans les réseaux de neurones, nécessaire pour résoudre des problèmes complexes.

Si nous traçons les sorties non linéaires produites par les fonctions d’activation, nous obtiendrons une courbure. La pente de la courbe est utilisée pour calculer le gradient. Et le gradient nous aide à comprendre le taux de changement et les relations entre les variables.

À partir des relations, les algorithmes sont optimisés et les poids sont mis à jour.

Il existe plusieurs fonctions d'activation :

![activation_functions](https://miro.medium.com/max/1400/1*ZafDv3VUm60Eh10OeJu1vw.png)


* La fonction <u>Rectified Linear Unit (ReLU)</u> est la fonction d’activation la plus simple et la plus utilisée : Elle donne x si x est supérieur à 0, 0 sinon.
* La fonction <u>Sigmoïde</u> donne une valeur entre 0 et 1, une probabilité. Elle est donc très utilisée pour les classification binaire, lorsqu’un modèle doit déterminer seulement deux labels.
* La fonction <u>Softmax</u> permet elle de transformer un vecteur réel en vecteur de probabilité. On l’utilise souvent dans la couche finale d’un modèle de classification, notamment pour les problèmes multiclasse.
* La fonction <u>tanh</u> est simplement la fonction de la tangente hyperbolique. Il s’agit en fait d’une version mathématiquement décalée de la fonction sigmoïde : sigmoïde donne un résultat entre 0 et 1, tanh donne un résultat entre -1 et 1. L’avantage de tanh est que les entrées négatives seront bien répertoriées comme négatives là où, avec sigmoïde, les entrées négatives peuvent être confondus avec les valeurs proche de nulles. Cette fonction est, comme Sigmoïde, utilisé dans la classification binaire.
* La fonction <u>Exponential Linear Unit (ELU)</u> est une amélioration de ReLU car elle permet d’avoir des valeurs lisses lorsque x < 0. Lorsque x < 0, ELU a des valeurs négatives différents de 0 (ce qui n’est pas le cas de ReLU). Cela permet de rapprocher la moyenne de la fonction de zéro.
* ...
* Il est possible de créer une fonction d'activation personnalisée


### La fonction de perte

Comme nous en avons déjà parlé plusieurs fois, la fonction de perte est la représentation mathématique permettant de mesurer la différence entre la réalité et la prédiction du modèle.

Entrainer un modèle correspond donc à minimiser cette fonction : c'est à dire trouver les paramètres permettant d'avoir la différence la plus faible possible entre la réalité et ce que prévoit le modèle. Cela semble logique...

Il existe ainsi plusieurs fonction de perte, qui seront utilisées dans différentes situations !

|Problématique            |Fonction de perte	                 |Fonction TensorFlow         |
|:-----------------------:|--------------------------------------|----------------------------|
|Classification binaire   |Entropie croisée binaire              |`binary_crossentropy`       |
|Multi classification     |Entropie croisée catégorique          |`categorical_crossentropy`  |
|Problème de régression   |Erreur absolue moyenne                |`mean_absolute_error ou mae`|
|Problème de régression   |Erreur quadratique moyenne            |`mean_squared_error ou mse` |
|Cas spécifique           |Question spécifique                   |**Fonction personnalisée**  |



### L'optimiseur

Vient maitenant le moment de minimiser la fonction de perte. Pour cela, nous faisons appel à un optimiseur.

Comme nous l'avons déjà vu, une manière possible de minimiser une fonction de perte, c'est de passer par la descente de gradient. Cela fonctionne très bien lorsque que la fonction de perte est convexe. Alors la descente de gradient, accompagnée du bon *learning rate* permettra d'atteindre le minimum de la fonction.

Cependant, comme nous l'avons également déjà dit, les fonctions de perte ne sont pas si souvent convexes... dans ce cas, il faut bien trouver une stratégie efficace.

C'est ainsi qu'ont été développé <u>divers optimiseurs</u> présentés [ici par exemple](https://www.analyticsvidhya.com/blog/2021/10/a-comprehensive-guide-on-deep-learning-optimizers/). A la fin de l'article, vous trouverez notamment un comparatif pour un même réseau de différents optimiseurs avec leurs performances aucours des entrainements et la durée des entrainements.

Synthétiquement nous pouvons notamment retrouver les optimiseurs :
* Gradient Descent
* Stochastic Gradient Descent
* Stochastic Gradient descent with momentum
* Mini-Batch Gradient Descent
* Adagrad
* RMSProp
* AdaDelta
* Adam

Communément, l'optimiseur **<u>Adam</u>** est souvent considéré comme le plus efficace. Il est souvent préféré et les tests commencent souvent par là.


## Entrainement par batch

Contrairement aux modèles classiques de Machine Learning, l'entraînement du modèle doit se faire ici par **"batchs"** : On divise la base de données d'entraînement en plusieurs batchs (ou lots) de la taille que l'on souhaite sur lesquels nous effectuons la descente de gradient. 

L'entraînement par batchs est nécessaire pour contourner les problèmes de mémoire des ordinateurs que nous utilisons de nos jours.

Lorsque toutes les données ont été utilisées, on dit que l'entraînement a complété une "epoch". On peut choisir le nombre d'epochs sur lesquelles le modèle va s'entraîner.
Par exemple, si nous choisissons de faire notre entraînement sur des batchs de taille 200 alors que la base de données contient 2000 entrées, chaque epoch consistera à faire la descente de gradient sur 2000/200 = 10 batchs.

Malheuresement, il n'y a pas de moyen analytique permettant de déterminer à l'avance le nombre d'epochs, l'optimizer ou la taille de batch qui donnera le meilleur résultat.
Pour l'instant, la méthode la plus efficace pour choisir est d'entraîner le modèle plusieurs fois en faisant varier un de ces hyperparamètres à la fois et de choisir la combinaison de paramètres qui donne le meilleur résultat sur l'échantillon de validation....



## Les différent type de couches

Comme indiqué précédemment, dans un réseau, les neurones sont regroupés en couches : la couche d'entrée, la couche de sortie et les couches cachées. Chaque couche applique une transformation sur chacun des éléments du tenseur d’entrée.

La principale différence entre les types de couches repose sur la manière dont les neurones se comportent.

### Couches dense :

La couche Dense est la couche de base en Deep Learning. Chaque neurone de cette couche possède un vecteur de poids de la même taille que le vecteur qui lui est passé en entrée et un terme de biais.

La couche dense est essentiellement utilisée pour modifier les dimensions du tenseur.

Les opérations effectuées par cette couche sont:
* Produit scalaire entre les vecteur de poids de chaque neurone et le vecteur d'entrée auquel on ajoute un terme de biais spécifique à chaque neurone.
* Activation des résultats des produits scalaires grâce par une fonction d'activation non-linéaire telle que  ReLU, tanh ou sigmoid.
* Concaténation des activations pour former un nouveau vecteur qui sera passé en entrée de la couche suivante.


### Couche de Convolution
La couche de Convolution est un peu plus complexe. Elle est principalement utilisée pour les tâches de Vision par Ordinateur, pour analyser des images.

![conv_layer](https://i0.wp.com/analyticsindiamag.com/wp-content/uploads/2018/01/conv-full-layer.gif?resize=543%2C544&ssl=1)

La convolution c’est le fait d’analyser une image ou une donnée par morceaux : il s'agit donc d'extraire des caractéristiques remarquables locales d'une image. Chacune de ces caractéristiques remarquable pourrait donc devenir la "feature" phare permettant de classifier.

Cette couche va réaliser le produit de convolution qui est une opération qui ressemble beaucoup au produit scalaire, sauf qu'il ne s'effectue pas sur un vecteur mais sur une matrice:
    D'un côté nous avons une matrice d'entrée, souvent appelée tuile ou convolution patch en anglais, et d'un autre nous avons une matrice de convolution, souvent appelée noyau de convolution, filtre ou convolution kernel en anglais. Ces deux matrices doivent absolument avoir les mêmes dimensions pour calculer leur produit de convolution.
    
Ce type de couche est particulièrement efficace pour différencier exrtraire les caractéristiques locales d'une image.

![result_conv](https://miro.medium.com/max/1400/1*wju0Urp6KpAT11wktTp5Sg.png)

### Couche de Pooling
Le Pooling est utilisé pour réduire la taille des données.

En effet, elle compresse le tenseur d’entrée pour n’en garder que l’information pertinente.

Comme la couche de Convolution, elle fonction par morceaux. Cependant la couche de Pooling va uniquement garder l’information importante de ce morceau et supprimer le reste.

Il existe différent Pooling comme :
* le Max Pooling, qui prend la valeur maximum dans un sous-groupe de données
* l’Average Pooling, qui prend la moyenne des valeurs dans un sous-groupe de données

Exemple d'opérations d'une couche de Max pooling : 
* La matrice de sortie d'une couche est découpée en plusieurs petites tuiles contenant des valeurs de la matrice.
* De chaque tuile on extrait la valeur maximale.
* On recompose une matrice ne contenant que les valeurs maximales de chaque tuile et on l'envoie dans la couche suivante.

![exemple_max-pool](https://nico-curti.github.io/NumPyNet/NumPyNet/images/maxpool.gif)


### Couche de Normalisation
La couche de Normalisation est sûrement la plus facile à comprendre.

Après l’utilisation d’une couche de Deep Learning, les données sont transformées. Cela à parfois pour effet de produire des valeurs dispersées.

Prenons un exemple... imaginons que certains neurone aient pour valeur de sortie 3 et d’autres 2867. C’est une échelle de valeur très dispersée et ce n’est pas bon pour notre modèle de Deep Learning. La couche de Normalisation applique donc une normalisation à ces données. On peut imaginer normaliser nos données pour qu’ils soient tous dans une échelle de 0 à 1. Par exemple, notre 3 deviendrait 0.04 et le 2867 serait 0.89.

Appliquer une normalisation à nos données après chaque couche permet au modèle d’apprendre plus facilement en faisant moins d’erreur.


### Couche de Régularisation
La couche de régularisation sert aussi à améliorer l’apprentissage du modèle.

Avec la normalisation, on simplifiait la tâche de notre modèle. Avec la Régularisation, on va la complexifier.

Effectivement, la Régularisation, par exemple la couche de **Dropout**, va transformer certaines valeurs de notre tenseur en 0. On parle d'inactiver des neurones.

Cette technique va augmenter la difficulté pour notre modèle mais paradoxalement cela va le rendre plus performant. En effet, en transformant aléatoirement certaines valeurs en 0, on oblige notre modèle a être plus performant pour résoudre la tâche qu’on lui a fixé. Cela force le réseau à diversifier ces sources d'apprentissage.

# Lançons nous dans la création de notre modèle

Dans un premier temps, nous allons créer (reproduire en réalité...) un réseau de neurones très simple. Ce réseau de neurone a été initialement proposé par M. LeCun, un grand nom du deep learning, dont vous entendrez probablement parler souvent si vous vous lancez dans le deep...

Nous allons commencer par la création de l'architecture de notre réseau.

Nous allons construire notre modèle séquentiellement, c'est-à-dire que nous allons instancier notre modèle et ensuite y ajouter les couches les unes après les autres, depuis la couche d'entrée jusqu'à la couche de sortie.

La bibliothèque Keras/tensorflow nous facilite grandement les choses :
1. Instancier un modèle avec le constructeur Sequential.
2. Créer les couches qui composeront le modèle avec leur paramètres.
3. Ajouter les couches au modèle grâce à sa méthode add (réalisée sur le modèle créé à l'étape 1).

La création des couches doit se faire de manière spécifique :
- Concernant la première couche du modèle : elle doit contenir les dimensions du vecteur d'entrée avec le paramètre input_size. _Cette précision **n'est pas** nécessaire pour les couches suivantes_.
- Le nombre de neurones dans une couche se définit avec le paramètre `units =`.
- Il est important d'initialiser les poids de nos neurones. Pour cela, Nous pouvons utiliser le paramètre `kernel_initializer =`.
- Comme mentionné précédemment, nos neurones doivent posséder une fonction d'activation. Pour cela, il est possible : soit d'instancier une couche d'activation puis l'ajouter au modèle, soit de définir la fonction d'activation dans le paramètre `activation =` lors de la création de la couche.


> <u>Consignes :</u>
> * Créer `model_NN`, une instance `Sequential()`. Nous ne préciserons aucun paramètres ici.
>
>
> * Créer `first_layer_NN`, la couche d'entrée de notre modèle. Il s'agit d'une couche **`Dense()`**. Cette couche dense doit prendre en paramètres : 300 neurones (units), la dimension d'entrée correspond au nombre de pixels (souvenez vous de la variable `pixel_number` que nous avons créé juste avant), nous utiliserons une fonction d'activation `'tanh'`, et l'initialisation de notre kernel sera `'normal'`.
> * Créer `second_layer_NN`, la couche de sortie de notre modèle. Il s'agit d'une couche **`Dense()`**. Cette couche dense doit prendre en paramètres : le nombre neurones (units) doit être égal au nombre de classes que nous avons (souvenez vous de la variable `classe_number` que nous avons créé juste avant), nous utiliserons une fonction d'activation `'softmax'`, et l'initialisation de notre kernel sera `'normal'`.
>
>
> * Vient maitenant le moment d'ajouter les couches à notre modèle (instancié à la première étape : `model_NN`). Pour cela, utiliser la méthode `.add()`, et ajouter successivement les couches `first / second _layer_NN` que nous avons créé précédemment.
>
>
> * Afficher le résumé (la structure) du modèle que nous venons de créer. Pour cela, nous pouvons nous appuyer sur la méthode `.summary()` que possède le modèle que nous venons de créer.

````python
pixel_number = 28*28

print(classe_number)

model_NN = Sequential()
first_layer_NN = Dense(units=300, 
                       input_dim=pixel_number, 
                       activation="tanh", 
                       kernel_initializer ="normal")

second_layer_NN = Dense(units=classe_number,
                        activation="softmax",
                        kernel_initializer="normal")

model_NN.add(first_layer_NN)
model_NN.add(second_layer_NN)

print(model_NN.summary())
````

# Paramétrons l'entrainement

Félicitation ! Vous venez de définir votre réseau de neurones !

Il faut maintenant configurer son processus d'entraînement. Pour cela, nous pouvons utiliser la méthode `.compile()`de notre modèle. Nous pouvons notamment mentionner **trois arguments**:
* **<u>Un optimiseur</u>** (paramètre `optimizer =`) (utilisé pour minimiser la fonction de perte).
* **<u>Une fonction de perte</u>** (paramètre `loss =`) à optimiser (à l'aide de l'optimiseur).
* **<u>Une liste de métriques</u>** (paramètre `metrics =`) afin d'évaluer la performance de notre modèle pendant son entraînement. La métrique 'accuracy' est utilisée pour évaluer la précision de la classification.

> <u>Consignes :</u>
> * Compiler notre **model_NN** à l'aide de la méthode `.compile()`. Nous emploierons :

    - `'Adadelta'` pour l'optimiseur (==> `keras.optimizers.Adadelta()`)
    - la métrique sera `['accuracy']`

````python
model_NN.compile(loss='categorical_crossentropy',
                optimizer="Adadelta",
                metrics=["accuracy"])
````

Pour entrainer un modèle déjà défini, il s'agit silplement d'utiliser la méthode `.fit()` que possède notre modèle. Il faut (ou, il est possible de) renseigner plusieurs arguments :
* __Le jeu d'entrainement__
* __Les labels__ (la cible, la target...) d'entrainement
* la taille des batchs souhaitées (`batch_size =`)
* __le nombre d'épochs__ (`epochs =`) : 1 epoch correspond simplement au moment ou le modèle s'est entrainé sur l'ensemble des données fournies. Imaginons que nous ayons 2000 entrées au total, que nous souhaitions des batchs de 200. Chaque epoch correspondrait donc à l'entrainement de 10 batchs de 200 entrées.
* La validation de notre score d'entrainement (`validation_split =`). En effet, il est possible de spécifiquer que nous souhaitons que le modèle valide ses résultats, comme nous pourrions le faire avec un train_test_split. Il est donc possible d'indiquer un nombre en 0 et 1 qui correspondra donc à la fraction des données mises de cotées pour valider l'apprentissage des données. Attention, il ne faut utiliser ce paramètres que si vous avez suffisemment de données d'entrainement.
* Le verbose, qui permet de suivre l'entrainement de notre modèle.

# Créons une fonction 

Nous allons créer une fonction qui reprend ces dernières étapes et qui nous permettra donc de visualiser les performances des modèles que nous entrainons.

> <u>Consignes :</u>
> * Créer `analyze_model()` une fonction qui prend en arguments **model**, **X_test**, **y_test**, **train_hist** et **epochs_nb** (initialisé à 20). Cette fonction doit :
>     - Créer **score** : l'évaluation par la méthode `.evaluate()`, sur les données de **test**.
>     - Créer **erreur**, le pourcentage d'erreur réalisé par le modèle.
>     - Afficher la valeur de la perte sur le set de test et l'erreur du modèle sur le set de test.
>     - Créer **train_acc** correspondant à l'**accuracy** du set d'entrainement de notre modèle à travers les épochs à partir de l'argument `train_hist` de notre fonction.
>     - Créer **val_acc** correspondant à l'accuracy de la validation de notre modèle à travers les épochs à partir de l'argument `train_hist` de notre fonction.
>     - Créer `epo` correspondant à **epochs_nb+1**
>     - Créer une figure récapitulant pour **chaque époque** l'accuracy pour le set d'entrainement et le set de validation (pendant l'entrainement). N'oubliez pas les titres et légendes.

````python
def analyze_model(model, X_test, y_test, train_hist, epochs_nb = 20) :
    
    erreur_model, score_model = model.evaluate(X_test, y_test)
    
    erreur_model = round(erreur_model*100,2)
    score_model = round(score_model*100,2)
    
    print("Score : ", score_model)
    print("Erreur : ", erreur_model)
    
    train_acc = train_hist.history["accuracy"]
    val_acc = train_hist.history["val_accuracy"]
    
    epo = epochs_nb+1
    
    plt.plot(range(1,epo),train_acc);
    plt.plot(range(1,epo),val_acc);
    plt.title("Evolution de l'accuracy pour chaque epoch");
    plt.xlabel("epoch");
    plt.ylabel("accuracy");
    plt.xticks(range(1,epo))
    plt.legend(title="Légende" ,labels=["train","validation"]);
    
````

Les images dont nous disposons ont une dimension de 28 pixels x 28 pixels. Etant en nuances de gris, elles n'ont qu'une seule dimension en **profondeur** (dimension 3 pour une image en rgb).

Cette fois, nous allons aussi nous appuyer sur des **couches de convolutions**, dans lesquelles chaque neurone possède son propre noyau de convolution. En général, les couches de convolution sont **suivies par des couches de pooling** afin de réduire progressivement la dimension de l'entrée. Ceci permet à la fois de régulariser le modèle (en diminuant les risques d'overfitting) et d'améliorer la performance de l'algorithme en termes de temps de calcul et d'utilisation de mémoire.

Les couches de convolutionont comme constructeur : **`Conv2D()`** qui prend différents paramètres:
- `filters =` : le nombre de matrice de convolution dans la couche.
- `kernel_size =` : les dimensions du noyau de convolution.
- `padding =` : détermine si le noyau a le droit de dépasser le bord de l'image.
- `input_shape =` : dimensions de l'image d'entrée. Doit <u>absolument</u> être précisé dans la première couche.
- activation : la fonction d'activation.


Avant toute choses... il reste quand même nécessaire de formater nos données pour qu'elles puisse être comprises par les couches de convolutions. Nous allons donc à nouveau procéder à la préparation des données.
````python
import itertools

def print_nice_conf_matrix(y, cnf_matrix):
    '''
    Fonction dediee a rendre une jolie matrice de confusion a partir de :
        - y : un array contenant les classes pour la classification. 
               Sert uniquement a compter le nombre de classes.
        - cnf_matrix : la matrice de confusion qui nous interesse.
    '''
    classes = range(0,(len(np.unique(y))))
    
    plt.figure()
    
    plt.imshow(cnf_matrix, interpolation='nearest',cmap='Blues')
    plt.title("Matrice de confusion")
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes)
    plt.yticks(tick_marks, classes)
    
    for i, j in itertools.product(range(cnf_matrix.shape[0]), range(cnf_matrix.shape[1])):
        plt.text(j, i, cnf_matrix[i, j],
                 horizontalalignment = "center",
                 color = "white" if cnf_matrix[i, j] > ( cnf_matrix.max() / 2) else "black")
    
    plt.ylabel('Vrais labels')
    plt.xlabel('Labels prédits')
    plt.show()
````

# Le transfer learning

Le transfer learning, bien qu’étant une approche en plein essor (mis en avant notamment par Andrew Yan-Tak Ng en 2016), est une approche dont les fondements étaient déjà décrits en 1976. Le transfer learning répond donc notamment à deux problématiques grandissantes :
1. Les données disponibles sont grandissantes;
2. Les ressources informatiques sont limitées alors que la complexité des modèles requiert des puissances de calculs toujours plus importantes.

Ainsi, le _“transfer learning”_ est une approche qui consiste à transférer à un nouveau modèle, le savoir d’un modèle déjà entraîné. En l’occurrence, il s’agit de figer les poids du premier modèle et de transférer ces poids (issus de la résolution d’un problème précis) afin de répondre à une nouvelle problématique résolue par un nouveau modèle.


De cette manière, le problème à résoudre bénéficie du résultat d’un entraînement préalable pour lequel nous n’aurions eu ni le temps, ni les ressources matérielles.


D’un point de vue pratique, le transfer learning s’implémente de manière relativement simple. Il convient d’importer un modèle pré-entraîné avec ses poids (en excluant les paramètres d’entrée pour pouvoir lui donner nos entrées spécifiques), il faut ensuite figer les poids de ce modèle puis l’inclure dans notre architecture de modèle en tant que modèle d’entrée.

Le transfer learning requiert tout de même deux problématiques proches l’une de l’autre afin de pouvoir s’appliquer.

Les modèles existants (VGG, ResNet, EfficientNet...) sont composés de deux grandes parties :
1. le skelette (backbone) : un ensemble de convolution permettant l'extraction des caractéristiques remarquables des images.
2. Une succession de couches **dense** qui ont pour objectif la classification.

Le transfer learning consiste donc à transférer le skelette principale d'un modèle déjà entrainé (en fait, les couches de convolutions principalement), puis ajouter nos propres couches de classification (des couches Dense).

## D'un point de vue pratique.... comment ça se passe ?

```python
    # On importe le modèle qui nous intéresse
from tensorflow.keras.applications import EfficientNetB1
    # On récupère notre modèle
efficientNet = EfficientNetB1(include_top=False, input_shape=(XX, XX, XX))
    # On fige les poids du modèle
for layer in efficientNet.layers:
    layer.trainable = False

    # Construction du modèle
model = Sequential()
model.add(efficientNet) # Après avoir instancié le modèle sequentiel on ajoute notre skelette
model.add(GlobalAveragePooling2D())
model.add(Dense(units = ............))
......
```

---

Lorsque nous avons importé le modèle déjà entrainé, nous avons réalisé 2 étapes :
1. Nous avons spécifié l'argument `include_top=False` ce qui nous permet de ne pas importer les couches **Dense** initialement entrainées dans le modèle pour la classification
2. Nous avons indiqué que nous ne souhaitions pas que le modèle puisse "apprendre" de nos images, nous avons laissé les poids des couches de convolution **figés** (`layer.trainable = False`)

Pour aller plus loin, et grapiller encore quelques points d'accuracy, il est possible de laisser la dernière ou les dernières couches de convolution apprendre de nos images directement. Dans ce cas, il faut simplement :

```python
efficientNet = EfficientNetB1(include_top=False, input_shape=(XX, XX, XX))

for layer in efficientNet.layers[-1:]:
    layer.trainable = True
```
